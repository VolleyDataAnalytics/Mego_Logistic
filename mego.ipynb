**Ant Colony Optimisation Algorithm**

```python
import numpy as np
import pandas as pd

# Load the distance matrix
file_path = 'distance_matrix.csv'  # Replace with your file path
distance_matrix = pd.read_csv(file_path, index_col=0)

# Convert distance matrix to a numpy array and handle infinities for self-distances
distance_matrix_values = distance_matrix.values
np.fill_diagonal(distance_matrix_values, np.inf)
distance_matrix_values[distance_matrix_values == 0] = np.inf

# Define ACO parameters
num_ants = 10
num_iterations = 100
alpha = 1
beta = 2
evaporation_rate = 0.5
pheromone_deposit = 100
num_destinations = distance_matrix_values.shape[0]

# Initialize pheromone and heuristic matrices
pheromone_matrix = np.ones_like(distance_matrix_values)
heuristic_matrix = 1 / distance_matrix_values

# Function to calculate transition probabilities
def calculate_transition_probabilities(pheromone_row, heuristic_row, visited, alpha, beta):
    probabilities = np.zeros_like(pheromone_row)
    for i in range(len(probabilities)):
        if i not in visited:
            probabilities[i] = (pheromone_row[i] ** alpha) * (heuristic_row[i] ** beta)
    probabilities /= probabilities.sum() if probabilities.sum() > 0 else 1
    return probabilities

# Function to calculate route distance
def calculate_route_distance(route, distance_matrix):
    return sum(distance_matrix[route[i], route[i + 1]] for i in range(len(route) - 1))

# Ant Colony Optimization function
def ant_colony_optimization(pheromone_matrix, heuristic_matrix, num_ants, num_iterations, alpha, beta, evaporation_rate):
    best_distance = np.inf
    best_route = None

    for iteration in range(num_iterations):
        all_routes = []
        all_distances = []

        for ant in range(num_ants):
            visited = [0]  # Start at Mumbai Dock (index 0)
            while len(visited) < num_destinations:
                current_city = visited[-1]
                probabilities = calculate_transition_probabilities(
                    pheromone_matrix[current_city], heuristic_matrix[current_city], visited, alpha, beta
                )
                next_city = np.random.choice(range(num_destinations), p=probabilities)
                visited.append(next_city)
            visited.append(0)  # Return to Mumbai Dock

            distance = calculate_route_distance(visited, distance_matrix_values)
            all_routes.append(visited)
            all_distances.append(distance)

        # Update pheromones
        pheromone_matrix *= (1 - evaporation_rate)
        for route, distance in zip(all_routes, all_distances):
            for i in range(len(route) - 1):
                pheromone_matrix[route[i], route[i + 1]] += pheromone_deposit / distance

        # Track the best route
        min_distance = min(all_distances)
        if min_distance < best_distance:
            best_distance = min_distance
            best_route = all_routes[all_distances.index(min_distance)]

    return best_route, best_distance

# Run ACO
best_route, best_distance = ant_colony_optimization(
    pheromone_matrix, heuristic_matrix, num_ants, num_iterations, alpha, beta, evaporation_rate
)

# Map indices back to destination names
best_route_names = [distance_matrix.index[i] for i in best_route]

# Output results
print("Best Route:", " -> ".join(best_route_names))
print("Best Distance:", best_distance)
```

**Genetic Algorithm**

```python
import numpy as np
import random
import pandas as pd

# Load the distance matrix file
distance_matrix = pd.read_csv('distance_matrix.csv', index_col=0)

# Convert to NumPy array
distance_array = distance_matrix.to_numpy()

# Parameters for the genetic algorithm
POPULATION_SIZE = 200
NUM_GENERATIONS = 100
MUTATION_RATE = 0.5
NUM_LOCATIONS = distance_array.shape[0]

# Generate an initial population with Mumbai as the start and end point
def initialize_population(pop_size, num_locations):
    population = []
    for _ in range(pop_size):
        route = random.sample(range(1, num_locations), num_locations - 1)  # Exclude Mumbai (index 0)
        route = [0] + route + [0]  # Add Mumbai as start and end
        population.append(route)
    return population

# Fitness function: Calculate total distance of a route
def calculate_fitness(route):
    total_distance = sum(
        distance_array[route[i], route[i + 1]] for i in range(len(route) - 1)
    )
    return 1 / total_distance  # Inverse of distance (higher fitness is better)

# Selection: Tournament selection
def tournament_selection(population, fitnesses):
    selected = random.sample(list(zip(population, fitnesses)), 5)
    return max(selected, key=lambda x: x[1])[0]

# Crossover: Ordered crossover (OX)
def ordered_crossover(parent1, parent2):
    size = len(parent1)
    start, end = sorted(random.sample(range(1, size - 1), 2))  # Exclude start and end points (Mumbai)
    child = [-1] * size
    child[start:end] = parent1[start:end]

    ptr = end
    for gene in parent2:
        if gene not in child:
            if ptr == size - 1:
                ptr = 1
            child[ptr] = gene
            ptr += 1

    child[0], child[-1] = 0, 0  # Ensure start and end at Mumbai
    return child

# Mutation: Swap mutation (excluding Mumbai)
def mutate(route, mutation_rate):
    if random.random() < mutation_rate:
        i, j = random.sample(range(1, len(route) - 1), 2)  # Exclude start and end points
        route[i], route[j] = route[j], route[i]
    return route

# Main genetic algorithm function
def genetic_algorithm():
    population = initialize_population(POPULATION_SIZE, NUM_LOCATIONS)

    for generation in range(NUM_GENERATIONS):
        fitnesses = [calculate_fitness(route) for route in population]
        new_population = []

        for _ in range(POPULATION_SIZE):
            parent1 = tournament_selection(population, fitnesses)
            parent2 = tournament_selection(population, fitnesses)
            child = ordered_crossover(parent1, parent2)
            child = mutate(child, MUTATION_RATE)
            new_population.append(child)

        population = new_population

        # Print the best route of the current generation
        best_fitness = max(fitnesses)
        best_route = population[fitnesses.index(best_fitness)]
        best_distance = 1 / best_fitness
        print(f"Generation {generation + 1}: Best Distance = {best_distance:.2f}")

    # Final output: Best route and its distance
    fitnesses = [calculate_fitness(route) for route in population]
    best_fitness = max(fitnesses)
    best_route = population[fitnesses.index(best_fitness)]
    best_distance = 1 / best_fitness
    return best_route, best_distance

# Run the genetic algorithm
best_route, best_distance = genetic_algorithm()

# Display the results
print("\nOptimized Route:", [distance_matrix.index[i] for i in best_route])
print("Total Distance:", best_distance)
```
